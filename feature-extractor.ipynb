```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import os
from tensorflow.keras.preprocessing import image
```

```python
MAX_SEQ_LENGTH = 55 # defining sequence 
NUM_FEATURES = 1024
IMG_SIZE = 128

EPOCHS = 50
```

```python
def build_feature_extractor():
    # Creates a feature extractor model that outputs 1024 features.
    feature_extractor = keras.applications.DenseNet121(
        weights="imagenet",
        include_top=False,
        pooling="avg",
        input_shape=(IMG_SIZE, IMG_SIZE, 3),
    )
    preprocess_input = keras.applications.densenet.preprocess_input

    inputs = keras.Input((IMG_SIZE, IMG_SIZE, 3))
    preprocessed = preprocess_input(inputs)

    outputs = feature_extractor(preprocessed)
    return keras.Model(inputs, outputs, name="feature_extractor")

feature_extractor = build_feature_extractor()

```

```python
def get_features_from_directory(directory_path):
    features = []
    labels = []
    
    for subdir, _, files in os.walk(directory_path):
        label = os.path.basename(subdir)  # Each subdirectory represents a class label
        for file in files:
            img_path = os.path.join(subdir, file)
            img = image.load_img(img_path, target_size=(IMG_SIZE, IMG_SIZE))
            img_array = image.img_to_array(img)
            img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
            
            # Preprocess and extract features
            feature = feature_extractor.predict(img_array)
            features.append(feature.flatten())  # Flatten the feature array
            labels.append(label)

    return np.array(features), np.array(labels)

# Example usage:
# Assuming images are organized in subfolders within 'data_directory'
data_directory = 'sample'
features, labels = get_features_from_directory(data_directory)

# # Initialize feature extractor

# Label preprocessing with StringLookup.
label_processor = keras.layers.StringLookup(
    num_oov_indices=0, vocabulary=np.unique(labels), mask_token=None
)
print(label_processor.get_vocabulary())
```
